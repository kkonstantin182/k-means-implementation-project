# Implementation of the K-means algorithm and its verification on "Customer Data"

```python
import pandas as pd
import numpy as np
from AddUtil import Preprocess, Vizualization, ClusterNumber
from Cluster import MyKMeans
from sklearn.cluster import KMeans
```

# 1. Data Processing

```python
path = "E:\\OneDrive\\Unimi\\Courses\\MyProjects\\data\\Mall_Customers.csv"
data = pd.read_csv(path)


# Let's perform clustering only on the two features.
# There are no missing values. Moreover, only only two features are used.
# That is why most of steps are omitted. 
# Only a feature scaling was perfromed.

Preprocess.drop_columns(data, columns=['CustomerID', 'Genre', 'Age']) 
X = Preprocess.get_features(data)
X = Preprocess.scale(X)
```

# 2. Results

## 2.1. Determining the optimal number of clusters

```python
ClusterNumber.elbow_method(MyKMeans(1), X, 10)
```

```python
ClusterNumber.elbow_method(KMeans(1), X, 10)
```

## 2.2. Clustering

```python
km1 = MyKMeans(5,seed=42)
km1.fit(X)
```

```python
km2 = KMeans(5, random_state=42)
km2.fit(X)
```

```python
# Let's visually compare methods from scikit-learn and the developed one.
# The label "Prediction" corresponds corresponds to the method presented in this work
# The label "Original" corresponds to the method implemented in scikit-learn
Vizualization.plot_clusters(X, km1, inverse_axes=True, true_labels=km2.labels_)
```
